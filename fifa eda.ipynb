## 2. Exploración básica de los datos.

### Estas son las variables por entrada que contiene este dataset.
+ **ID:** número de identificación del jugador. 
+ **Name:** nombre del jugador.
+ **Age:** edad del jugador.
+ **Nationality:** nacionalidad del jugador.
+ **Overall:** Habilidad general de cada jugador.
+ **Potential:** potencial.
+ **Club:** Equipo de fútbol.
+ **Value:** Valor del jugador.
+ **Wage:** Salario
+ **Preferred Foot:** zurdos / diestros.
+ **International Reputation:** reputación internacional del jugador.
+ **Skill Moves:** Habilidad de movimiento.
+ **Position:** Posición en el juego.
+ **Joined:** Entrada al equipo.
+ **Contract Valid Untill:** contrato válido hasta.
+ **Height:** altura.
+ **Weight:** peso.
+ **Release Clause:** Claúsula de liberación.

## 3. Gráficos.

### Los gráficos están basados en la nueva variable "contrato_vigente" que son los jugadores que poseen contrato vigente de 2023 en adelante.

```python
contrato_vigente["Overall"].plot(kind = 'hist')
plt.show()
```

Hay 250 jugadores logran marcar goles entre 63 y 65.
Los que logran marcar 50 goles no son buenos jugadores, ya que la frecuencia es poca, por el contrario, los jugadores que logran marcar 90 goles, son excepcionales.

```python
# Comparamos nacionalidades y habilidad.
sns.catplot(x="Nationality", y="Overall", data=real_madrid, kind="bar", aspect=1, height=6)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

La mejor marca de goles la tiene la nacionalidad belga.

```python
# Comparamos la nacionalidad con el potencial.
sns.catplot(x="Nationality", y="Potential", data=contrato_vigente, kind="bar", aspect=4, height=5)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

Se observa en la gráfica que están por encima de 50 pocas nacionalidades, de ellas la que más destaca es Cabo verde, de puede deber a que allía hay mejores entrenadores para los futbolistas o su condición física sea más favorable.

```python
sns.catplot(x="Nationality", y="Age", data=real_madrid, kind="bar", aspect=2, height=10)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

```python
# Mapa de calor
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(df.corr(), mask=mask, vmax=1., vmin=-1., center=0,
square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```
