```python
# Vemos el jugador más joven. Con loc podemos seleccionar variables.
df.loc[df['Age'] == df['Age'].min()]
```

```python
# Jugadores más joven, con mejor puntuación y nacionalidad.
df.loc[df['Age'] == df ['Age'].min()][['Name', 'Age', 'Overall', 'Nationality']]
```

```python
# Vemos el jugador con mas puntuación 
Puntuacion=df.loc[df['Overall'] == df ['Overall'].max()][['Name', 'Age', 'Overall', 'Nationality']]
Puntuacion
```

```python
# El jugador mas mayor tiene 45 años
edad = df.loc[df['Age'] == df ['Age'].max()]
edad
```

```python
# Miro los que se les ha caducado el contrato
df[df["Contract Valid Until"] < '2021-01-01']
```
