## 2. Exploración básica de los datos.

### Estas son las variables por entrada que contiene este dataset.
+ **ID:** número de identificación del jugador. 
+ **Name:** nombre del jugador.
+ **Age:** edad del jugador.
+ **Nationality:** nacionalidad del jugador.
+ **Overall:** total de goles.
+ **Potential:** potencial.
+ **Club:** Equipo de fútbol.
+ **Value:** Puntaje de los jugadores.
+ **Wage:** Salario
+ **Preferred Foot:** zurdos / diestros.
+ **International Reputation:** reputación internacional del jugador.
+ **Skill Moves:** Habilidad de movimiento.
+ **Position:** Posición en el juego.
+ **Joined:** Entrada al equipo.
+ **Contract Valid Untill:** contrato válido hasta.
+ **Height:** altura.
+ **Weight:** peso.
+ **Release Clause:** Claúsula de liberación.

```python
# Identificación de variables:
categoricas = ["Name" , "Nationality", "Club", "Preferred Foot", "Position", "Contract Valid Until"]
numericas = ["ID" , "Age", "Overall", "Potential", "Joined", "Value", "Wage", "Internation Reputation", "Skill Moves", "Weight", "Height", "Release Clause" ]
```

```python
#Información por columna, moda.
for col in categoricas:
    print(f"Column name: {col}")
    print(df[col].value_counts())
    print()
```

Se observa:
+ No hay datos duplicados por lo que se supone que pueden ser homónimos o falta información en "name".

```python
# Se reducen variables que no se tendrán en cuenta al no estar especificadas.
df.drop(["ID", "Skill Moves", "Joined", "Height", "Weight", "Release Clause", "Preferred Foot"], axis = 1, inplace = True)
```

```python
# Se comprueba la eliminación de variables.
df.shape
```

```python
# Se muestran las columnas con las que se van a trabajar.
df.columns
```

```python
# Buscamos a los jugadores más jóvenes. Con ".loc" podemos seleccionar variables, con .min() encontramos los valores más bajos.
df_jovenes = df.loc[df['Age'] == df['Age'].min()]
df_jovenes
```

Se observa:
+ Los jugadores más jóvenes son de 16 años. 
+ La mayoría son diestros.
+ El nivel de reputación de todos es de 1.0.
+ Se unieron entre los años 2017 y 2018.

```python
# Jugadores más jóvenes, con mejor marca de goles y nacionalidad.
df.loc[df['Age'] == df ['Age'].min()][['Name', 'Age', 'Overall', 'Nationality']]
```

Se observa:
+ Los jugadores más jóvenes con mejor marca de goles provienen de Francia y España.

```python
# Vemos a los jugadores con mejor marca de goles, edad y nacionalidad y salario.
Puntuacion=df.loc[df['Overall'] == df ['Overall'].max()][['Name', 'Age', 'Overall', 'Nationality','Wage']]
Puntuacion
```

Se observa:
+ L.Messi y Cristiano Ronaldo son los jugadores con mejor marca de goles
+ Siendo los 36 años la edad límite para seguir en la Fifa, ambos jugadores se encuentran a cinco y tres años de cumplirlos.
+ Son de diferentes nacionalidades.
+ El salario no especifica 

```python
# El jugador con mas edad y sus datos.
edad = df.loc[df['Age'] == df ['Age'].max()]
edad
```

```python
# Jugadores los que se les ha caducado el contrato
df[df["Contract Valid Until"] < '2021-01-01']
```

Los jugadores que se les ha rescindido el contrato son 10996.

```python
# Se eliminan a los jugadores con contrato vencido. Con .drop eliminamos entradas. Con .index extrae datos según columnas
contrato_caducado = df[df["Contract Valid Until"] < '2022-01-01'].index
contrato_caducado = df.drop(contrato_caducado)
contrato_caducado
```

Actualmente contamos con 2851 entradas y 18 columnas, que son los jugadores con contratos vigentes a 2022.

```python
# Se comprueba la cantidad de entradas y columnas.
contrato_caducado.shape
```

Se observa que se ha reducido a 2851 entradas.

```python
# Se renombra la variable para seguir trabajando con los jugadores con contrato vigente.
contrato_vigente=contrato_caducado[contrato_caducado["Contract Valid Until"] >= '2023-01-01']
contrato_vigente
```

Actualmente tenemos 1085 y 18 columnas, que son los jugadores con contratos vigentes a 2023 en adelante.

```python
# Jugadores que más ganan 
contrato_vigente.loc[contrato_vigente['Overall'] == contrato_vigente ['Overall'].max()][['Name', 'Age', 'Wage', 'Nationality']]
```

```python
# Se hace una variable en la que nos muestre los jugadores con contrato vigente del Real Madrid.
real_madrid = contrato_vigente[contrato_vigente["Club"] == "Real Madrid"]
print(real_madrid.shape)
```

Nos muestra que son 6 jugadores con contrato vigente que juegen en el Real Madrid.

## 3. Gráficos.

### Los gráficos están basados en la nueva variable "contrato_vigente" que son los jugadores que poseen contrato vigente de 2023 en adelante.

```python
contrato_vigente["Overall"].plot(kind = 'hist')
plt.show()
```

Hay 250 jugadores logran marcar goles entre 63 y 65.
Los que logran marcar 50 goles no son buenos jugadores, ya que la frecuencia es poca, por el contrario, los jugadores que logran marcar 90 goles, son excepcionales.

```python
sns.catplot(x="Nationality", y="Overall", data=real_madrid, kind="bar", aspect=1, height=6)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

```python
# Comparamos la nacionalidad con el potencial.
sns.catplot(x="Nationality", y="Potential", data=contrato_vigente, kind="bar", aspect=4, height=5)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

Se observa en la gráfica que están por encima de 50 pocas nacionalidades, de ellas la que más destaca es Cabo verde, de puede deber a que allía hay mejores entrenadores para los futbolistas.

```python
sns.catplot(x="Nationality", y="Age", data=real_madrid, kind="bar", aspect=2, height=10)
plt.xticks(rotation=90)
plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

```python
sns.scatterplot(data=barcelona, x="Age", y="Potential", hue="Nationality")
```
